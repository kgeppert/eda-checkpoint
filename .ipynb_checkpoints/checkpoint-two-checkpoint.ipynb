## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#want to compare yield per colony between states from a specific region
#need to create a new df that is specific to the question
colony_yield_state_df = honey_df[honey_df['State'].isin(['Illinois', 'Indiana', 'Michigan', 'Wisconsin', 'Ohio'])].drop(['Honey producing colonies (thousand)', 'Stocks December 15  (1,000 pounds) ', 'Production (1,000 pounds)', 'Average price per pound (dollars)', 'Value of production (1,000 dollars)'], axis = 1)
colony_yield_state_df
```

```python
#using df.pivot instead of pd.pivot_table made sure the index was 'Year'
gl_yield = colony_yield_state_df.pivot(index = 'Year', columns = 'State', values = 'Yield per colony (pounds)')
gl_yield
```

```python
# chart using matplotlib
gl_yield.plot(kind='line')
```

```python
#chart using seaborn; did not need to pivot colony_yield_state_df in order to get chart to plot like I wanted it to. 
yield_chart = sam.lineplot(
    x = 'Year',
    y = 'Yield per colony (pounds)',
    hue = 'State',
    data = colony_yield_state_df
).set_title('Yield from 2015-2020 by State')
plt.show
```
